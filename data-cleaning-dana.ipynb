```python
# Import packages
import pandas as pd
import numpy as np
```

```python
# Read excel file
df=pd.read_excel('ncsr.xlsx')
# Cut off bottom
df=df[:len(df)-7]
df.tail()
```

```python
# Variables assigned
features=['D26I','D26K','D26M','D26N','D26O','D26Q','D26V','D26DD','D26EE']
```

```python
# Replace all 5's with 0's
df[features]=df[features].replace(5, 0)
```

```python
# Replace all negative numbers with NaN
df[features]=df[features].replace([-7,-8,-9], np.NaN)
```

```python
for name in features:
    df[name].fillna(df[name].mean())
```

```python
# Save to excel file
df[features].to_excel('ncsr_clean_dana.xlsx')
```
